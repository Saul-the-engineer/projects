# Fine-tuning llama 2 with Diriect Preference Optimization (DPO)

RLHF pipeline consists of 3 phases:
1. Supervised Fine-tuning (SFT)
   - On high-quality datasets on tasks of interests such as dialogue, summarization ect
2. Preference Sampling (PS)
   - SFT model is prompted with prompts x to produce pairs of answers. These are then presented to human labelers who express preferences for one answer over the other and reward learning.
3. Reinforcement Learning Optimization
   - We use the learned reward function to provide feedback to the language model. In particular we formulate an optimization problem and maximize using PPO.

References:
J. Schulman, F. Wolski, P. Dhariwal, A. Radford, and O. Klimov. Proximal policy optimization algorithms. arXiv preprint arXiv:1707.06347, 2017.

Direct Preference Optimization:
* Unlike prior RLHF methods which learn a reward and then optimize it via RL, DPO bypasses the reward modeling step and directly optimizes a language model using preference data.
* Leverage an anlalytical mapping from reward functions to optimal policies which enable us to transform a loss function over reward functions into a loss function over policies.
* This change of variables allows us to skip the explicit reward modeling step, while still optimizing under existing models of human preferences such as the Bradley-Terry model.




References:
R. Rafailov, A. Sharma, and E. Mitchel. Direct Preference Optimization: Your language model is secretly a reward model
