# State-Space Models

A state-space model (SSM) is a partially observed Markov Model, in which the hidden state, zt, evolves over time according to a Markov process. Each hidden state generates some observations yt at each time step. The main goal is to infer the hidden states given the observation. However, we may also be interested in using the model to predict future observations (e.g., for time-series forecasting or next-word prediction).

An SSM can be represented as a stochastic discrete time non-linear dynamical system of the form:



zt = f(zt-1, ut, qt)
yt = h(zt, ut, y1:t-1, rt)

Where:
- zt is the hidden state at time t
- ut are optional observed inputs
- yt are observed outputs
- f is the transition function
- q is the process noise
- h is the observation function
- r is the observation noise
